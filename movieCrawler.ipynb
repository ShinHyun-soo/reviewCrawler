```python
from urllib.request import urlopen
```

```python
from bs4 import BeautifulSoup
```

```python
review_list = []
for page in range(1,51):
  url = f'https://movie.naver.com/movie/bi/mi/pointWriteFormList.naver?code=154282&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false&page={page}'
  html = urlopen(url)
  soup = BeautifulSoup(html,'html.parser')
  for i in range(10):
    review = soup.find('span',{'id':f'_filtered_ment_{i}'})
    review = review.get_text().strip()
    review_list.append(review)
review_list
```
